```python
def calculate_distances(row):
    action_location = np.array([row['x0'], row['y0']])
    teammate_distances = []
    opponent_distances = []

    for player in row['freeze_frame']:
        player_location = np.array(player['location'])
        dist = distance.euclidean(action_location, player_location)

        if player['teammate']:
            teammate_distances.append(dist)
        else:
            opponent_distances.append(dist)

    # Calculate average distance of teammates and opponents from the action
    row['avg_teammate_distance'] = np.mean(teammate_distances) if teammate_distances else -1
    row['avg_opponent_distance'] = np.mean(opponent_distances) if opponent_distances else -1

    return row

event_df = event_df.apply(calculate_distances, axis=1)
```

```python
# count players
def count_players(row):
    teammate_count = 0
    opponent_count = 0

    for player in row['freeze_frame']:
        if player['teammate']:
            teammate_count += 1
        else:
            opponent_count += 1

    row['num_teammates'] = teammate_count
    row['num_opponents'] = opponent_count

    return row

event_df = event_df.apply(count_players, axis=1)
```
